```python
# y= b0 + b1* x1
#y-dependent variable 
# x-independent variable 
# b1-coefficient
# note for this no need for the feature scaling
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Salary_Data.csv')
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,1].values
```

splitting the dataset

```python
#spliting
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=1/3,random_state=0)
```

fitting the linear regression to the traning set


```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()#just keep the default
lr.fit(xtrain,ytrain)
```

predicting the test set results

```python
ypredicted=lr.predict(xtest)
```

visualizing the results

```python
# xaxis yaxis
plt.scatter(xtrain,ytrain,c='red')
plt.plot(xtrain,lr.predict(xtrain),c='blue')
plt.title('Train set')
plt.ylabel('salary')
plt.xlabel('year of experience')
```

```python
# test set
plt.scatter(xtest,ypredicted)
plt.plot(xtrain,lr.predict(xtrain))
plt.title('test set')
plt.xlabel('year of experience')
plt.ylabel('salary')
```

r2_score


```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(ytest, ypredicted)
mse = mean_squared_error(ytest, ypredicted)
r_squared = r2_score(ytest, ypredicted)
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r_squared)
```
